# Convert data to numpy arrays


```python
# Get data
Velocidad   = df.Velocidad.to_numpy().reshape(-1, 1)
Fuel1       = df.FUELS1.to_numpy().reshape(-1, 1)
Fuel2       = df.FUELS2.to_numpy().reshape(-1, 1)
TotalFuel   = df.TOTALFUEL.to_numpy().reshape(-1, 1)
#FechaGPS    = df["Fecha GPS"].to_numpy()
#RMP         = df.RPM.to_numpy()
```

# Exploratory Data Aanalysis (EDA)

```python
plt.figure(figsize=(10, 6))

plt.plot(Velocidad, label="Velocidad")
plt.plot(Fuel1, label="Tanque 1")
plt.plot(Fuel2, label="Tanque 2")

plt.legend(loc='best')

plt.ylabel("Amplitud")
plt.xlabel("Tiempo")
plt.title(url)

plt.show()
```
