# Axes labels,  legend, grid, text

```python
mes   = np.arange(1,13,1)


Tavg  = np.array([20,23,28,31,33,32,30,29,28,25,24,22])
Tmin  = Tavg - 6
Tmax  = Tavg + 6
```

```python
plt.plot(mes, Tavg, 'o:', label="Avg", markersize=8)
plt.plot(mes, Tmin, 's:', label="min", markersize=8)
plt.plot(mes, Tmax, 's:', label="max", markersize=8)

plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.title('Weather in Guadalajara')

plt.legend(loc='upper right', fontsize="small",shadow=False)

plt.grid(True)

plt.xlim(0, 13)
plt.ylim(10, 40)

plt.show()
```

# Xticks and Yticks

```python
mesname = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.plot(mes, Tavg, 'o:', label="Avg", markersize=8)
plt.plot(mes, Tmin, 's:', label="min", markersize=8)
plt.plot(mes, Tmax, 's:', label="max", markersize=8)

plt.xticks(mes,mesname)

plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.title('Weather in Guadalajara')

plt.legend(loc='upper right', fontsize="small", shadow=False)

plt.show()
```

# Adding a text

```python
plt.plot(mes, Tavg, 'o:', label="Avg", markersize=4)
plt.plot(mes, Tmin, 's:', label="min", markersize=4)
plt.plot(mes, Tmax, 's:', label="max", markersize=4)

plt.xticks(mes,mesname)

plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.title('Weather in Guadalajara')

plt.legend(loc='upper right', fontsize="large",shadow=False)

plt.grid(True)

plt.xlim(0, 13)
plt.ylim(10, 40)

plt.text(4, 15, 'Fancy plot', fontsize='xx-large')

plt.show()
```

# Subplots

```python
plt.subplot(2,1,1)
plt.plot(mes, Tmin, 's:', label="min", markersize=4)
plt.ylabel('Min temp (°C)')
plt.title('Weather in Guadalajara')

plt.subplot(2,1,2)
plt.plot(mes, Tmax, 's:', label="max", markersize=4)
plt.xlabel('Month')
plt.ylabel('Max temp (°C)')

plt.show()
```

# Scatter plot

Suppose that we have two data arrays g1 and g2 that we know they are somehow correlated; then, we can make a plot with g1 and g2 as the axis to see this correlation.

```python
x1 = np.random.normal(100, 10, size=800)
x2 = np.random.normal(200, 30, size=800)

plt.scatter(x1, x2, c='y', label='Old')

plt.xlabel("x1")
plt.ylabel("x2")

plt.show()
```

```python
# Let's generate a new pair of arrays, but with less dispersion,
#x1 = np.random.uniform(-10,10,size=1000)
x1 = np.random.normal(0,2,size=1000)
x2 = 2*x1 + np.random.normal(1,1,size=1000)

plt.scatter(x1, x2, c='g', label='Scatter plot')

plt.legend(loc='best')

plt.xlabel("x1")
plt.ylabel("x2")

plt.show()
```

```python
x1 = np.random.normal(100, 10, size=800)
x2 = np.random.normal(200, 30, size=800)

y1 = np.random.normal(110, 5, size=100)
y2 = np.random.normal(150, 10, size=100)

plt.scatter(x1, x2, c='r', label='New data', alpha=0.5)
plt.scatter(y1, y2, c='b', label='Old data', alpha=0.5)

plt.legend(loc='best')

plt.xlabel("x1")
plt.ylabel("x2")

plt.show()
```

# Reading text files and plotting its content

The most friendly format is text files, and the **Python** method to read these data is **loadtxt**

```python
data = np.loadtxt('DataFile.txt').T

print(type(data))
print(data.shape)

```

```python
plt.plot(data[0], label='Signal 1')
plt.plot(data[1], label='Signal 1')
plt.plot(data[2], label='Signal 1')

plt.legend(loc='best')

plt.xlabel("Amplitude")
plt.ylabel("Time")
plt.title("Signals")

plt.show()
```

```python
# Calcular the number of signals and samples in the data
Nsignals,Nsamples = data.shape

print("Number of signals: ", Nsignals)
print("Number of samples: ", Nsamples)

# Let's assume sampling frequency of 256 Hz
fs = 256

# Now, we can compute a time vector
t  = np.arange(Nsamples)/fs
```

```python
plt.plot(t,data[0], label='Signal 1')
plt.plot(t,data[1], label='Signal 1')
plt.plot(t,data[2], label='Signal 1')

plt.legend(loc='best')

plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
plt.title("Signals")

plt.show()
```

# Plot with error bars

```python
t  = np.arange(0,30,1) 
ft = 10*np.sin(0.1*t + np.pi/3)

plt.plot(t,ft,'o:')

plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
plt.title("Data with error bars")

plt.show()
```

```python
fterr = np.random.normal(1,0.5,size=len(t))

plt.errorbar(t,ft,fterr)

plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
plt.title("Data with error bars")

plt.show()
```

# Save data in a .txt file

We can save our data in a .txt file

```python
data2txt = t,ft,fterr # data2txt is a tuple, of numpy arrays, check by yourself

np.savetxt("DataFile_Saved.txt",data2txt)
```
